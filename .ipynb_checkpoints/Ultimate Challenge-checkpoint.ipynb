<h2>Part 1 - Exploratory data analysis</h2>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
logins_df = pd.read_json('logins.json')
logins_df.info()
logins_df.describe()
```

<b>There does not appear to be any NaN or null data.  All login_time data samples were converted to datetime64 data type.</b>

```python
time = pd.DatetimeIndex(logins_df.login_time)
logins_df['number_of_minutes'] = (time.hour * 60) + time.minute + (time.second / 60)
logins_df.loc[logins_df.number_of_minutes > 1440].number_of_minutes.value_counts()
```

```python
minute_bins = [i for i in np.arange(0, 1440, 15)]
minute_bins.append(np.inf)
bins = [*range(0, len(minute_bins) - 1)]
logins_df['minute_interval'] = pd.cut(
    logins_df.number_of_minutes,
    minute_bins,
    labels=bins
).astype('float')

logins_df.head()
```

```python
grouped_logins = logins_df.groupby(by='minute_interval').count()
grouped_logins.rename(columns={'login_time':'login_count'}, inplace=True)
grouped_logins.plot(y='login_count', title='Ultimate user login count over 15 minute intervals')
```

<b>As the login_count plot shows, there are peaks in demand around midday and midnight. Late morning has the lowest demand, and afternoon/evening has moderate demand.</b>

<h2>Part 2 - Experiments and metrics design</h2>

<b>1.  Key measure of success is to look at the number/proportion of passengers picked up in both cities for each driver.  A comparison would be made between a select group of drivers that is offered the toll reimbursement and the rest of the drivers.</b>

<b>2.  a. Implmentation - Notify randomly selected Gotham/Metropolis drivers that Ultimate is offering to reimburse them for toll bridge fees for a limited time.  A toll tracking feature can be added to the Ultimate app to determine when the drivers use the toll bridge.  That usage can be added to time series data in the form of a flag (1 for used toll bridge for the day, 0 for did not use the toll bridge for the day), or a flag could be added to all drivers' accounts signifying whether reimbursement is activated or not.  Presumably location of passenger pickups is already being recorded so the number of passengers picked up for each driver per day in each city could be calculated.

b. Testing - A few paired tests could be done to determine if the toll reimbursement has any effect number of passengers picked up in each city:  
    <p>Test 1 - t-test to compare the mean daily total number of passengers picked up in Gotham between toll reimbursed and non-toll reimbursed Gotham drivers.</p>
    <p>Test 2 - t-test to compare the mean daily total number of passengers picked up in Metropolis between toll reimbursed and non-toll reimbursed Gotham drivers.</p>
    <p>Test 3 - t-test to compare the mean daily number of passengers picked up in Gotham between toll reimbursed and non-toll reimbursed Metropolis drivers.</p>
    <p>Test 4 - t-test to compare the mean daily number of passengers picked up in Metropolis between toll reimbursed and non-toll reimbursed Metropolis drivers.</p>

c. Results - If 2 or more of the paired tests show differences in mean daily total number of passengers picked up, then we confidently say that the toll reimbursement policy is having a significant effect on at least one city for driver availability.  If 1 or more tests in each city shows differences in mean daily total number of passengers picked up, then we can confidently say that the toll reimbursement policy is having a significant effect on both cities.
</b>



<h2>Part 3 - Predictive modeling</h2>

<b>Let's load and take an overall look at the data in the ultimate_data_challenge.json file.</b>

```python
ultimate_df = pd.read_json('ultimate_data_challenge.json')
ultimate_df.info()
ultimate_df.describe()
```

<p><b>There are some missing values for avg_rating_of_driver, phone, and avg_rating_by_driver.  Not many missing values for phone, so we can just drop those entries from the dataframe.  One possible explanation for so many missing values for avg_rating_of_driver and avg_rating_by_driver is that passengers and drivers just never got around to submitting those ratings.  That's common for a lot of apps.</b></p>
<p><b>Now let's explore any outliers.  According to the quartile table above there are many high outliers for trips_in_first_30_days, avg_sruge, surge_pct, avg_dist.  Box plots will show how far above the distribution those outliers are.</b></p>

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(20, 20))

ax1.boxplot(ultimate_df.trips_in_first_30_days)
ax1.title.set_text('Outliers for trips in first 30 days')
ax2.boxplot(ultimate_df.avg_surge)
ax2.title.set_text('Outliers for average surge multiplier')
ax3.boxplot(ultimate_df.surge_pct)
ax3.title.set_text('Outliers for % of trips with surge multiplier > 1')
ax4.boxplot(ultimate_df.avg_dist)
ax4.title.set_text('Outliers for average distance in miles per trip for first 30 days')
```

<b>Drop those outliers.</b>

```python
ultimate_df = ultimate_df.loc[
    (ultimate_df.trips_in_first_30_days < 60) &
    (ultimate_df.avg_surge < 5) &
    (ultimate_df.surge_pct < 90) &
    (ultimate_df.avg_dist < 80)
]
```

<b>From the histograms below of the features that were just box-plotted, we can't quite tell that these are normal distributions due to long tails at the high end.  We can keep those data points in for now.</b>

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(20, 20))

ax1.hist(ultimate_df.trips_in_first_30_days)
ax1.title.set_text('Distribution for trips in first 30 days')
ax2.hist(ultimate_df.avg_surge)
ax2.title.set_text('Distribution for average surge multiplier')
ax3.hist(ultimate_df.surge_pct)
ax3.title.set_text('Distribution for % of trips with surge multiplier > 1')
ax4.hist(ultimate_df.avg_dist)
ax4.title.set_text('Distribution for average distance in miles per trip for first 30 days')
```

<b>Remove entries with missing phone values.</b>

```python
ultimate_df = ultimate_df.loc[pd.notna(ultimate_df.phone)]
```

<b>Translate date data into a format that is easier to use in pandas.</b>

```python
ultimate_df['last_trip_date'] = pd.to_datetime(ultimate_df.last_trip_date, infer_datetime_format=True)
ultimate_df['signup_date'] = pd.to_datetime(ultimate_df.signup_date, infer_datetime_format=True)
ultimate_df.head()
```

<b>Find the min and max last_trip_date months, and generate target labels from that data.  Any last trip date in June or later with signup_date at least 2 months prior can be treated as user that will be active in 6 months.</b>

```python
print(ultimate_df['last_trip_date'].max().month)
print(ultimate_df['last_trip_date'].min().month)
```

```python
ultimate_df['six_mth_active'] = np.where(
    (ultimate_df.last_trip_date.dt.month > 5) & 
    (ultimate_df.signup_date.dt.month < 5),
    1, 0)
ultimate_df['ultimate_black_user'] = np.where(ultimate_df.ultimate_black_user, 1, 0)
ultimate_df['has_rating_of_driver'] = np.where(pd.notna(ultimate_df.avg_rating_of_driver), 1, 0)
ultimate_df['has_rating_by_driver'] = np.where(pd.notna(ultimate_df.avg_rating_by_driver), 1, 0)
```

<b>Categorize city and phone in numerical IDs.</b>

```python
# (0: Astapor, 1: King's Landing, 2: Winterfell)
# (0: Android, 1: )
ultimate_df['city_id'] = ultimate_df.city.astype('category')
ultimate_df['city_id'] = ultimate_df.city_id.cat.codes
ultimate_df['phone_id'] = ultimate_df.phone.astype('category')
ultimate_df['phone_id'] = ultimate_df.phone_id.cat.codes
```

<b>Have a secondary data set where users that are missing avg_rating_of_driver or avg_rating_by_driver are dropped.  That way we can experiment with rating scores as features.</b>

```python
reduced_ult_df = ultimate_df.loc[(pd.notna(ultimate_df.avg_rating_of_driver)) & (pd.notna(ultimate_df.avg_rating_by_driver))]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="white")
fig, ax = plt.subplots(figsize=(11, 9))
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(ultimate_df.corr(), cmap=cmap, vmax=1, vmin=-1, center=0, square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

<b>From the correlation matrix we see that avg_surge and surge_pct are highly correlated, so we only need one of them as part of the feature set.  We will keep surge_pct.</b>

<b>The feature superset is small overall, we will skip Kbest analysis and just try different combinations of features to feed to the model(s).  Since there is a large range of values among the different features, we will need to scale all features to between 0 and 1.  This will ensure that Logistic Regression will converge.  Scaled data does not need to be used for Gradient Boosting.</b>

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression
from xgboost.sklearn import XGBClassifier

from sklearn.metrics import recall_score, precision_score, roc_curve

# feature_list = [
#     'trips_in_first_30_days', 'surge_pct', 'ultimate_black_user',
#     'weekday_pct', 'avg_dist', 'city_id', 'phone_id', 'has_rating_of_driver', 'has_rating_by_driver'
# ]
# feature_set = ultimate_df[feature_list]
# X_train, X_test, y_train, y_test = train_test_split(
#     feature_set, ultimate_df['six_mth_active'], test_size=.25, random_state=42, shuffle=True
# )

# feature_list = [
#     'trips_in_first_30_days', 'surge_pct', 'ultimate_black_user',
#     'weekday_pct', 'avg_dist', 'city_id', 'phone_id', 'avg_rating_of_driver', 'avg_rating_by_driver'
# ]
# feature_set = reduced_ult_df[feature_list]
# X_train, X_test, y_train, y_test = train_test_split(
#     feature_set, reduced_ult_df['six_mth_active'], test_size=.25, random_state=42, shuffle=True
# )

feature_list = [
    'trips_in_first_30_days', 'surge_pct', 'phone_id', 'city_id', 'ultimate_black_user', 'weekday_pct'
]
feature_set = ultimate_df[feature_list]
X_train, X_test, y_train, y_test = train_test_split(
    feature_set, ultimate_df['six_mth_active'], test_size=.25, random_state=42, shuffle=True
)

scaler = MinMaxScaler()
X_train_scale = scaler.fit_transform(X_train.values)
X_test_scale = scaler.fit_transform(X_test.values)
```

```python
clf = LogisticRegression()

clf.fit(X_train_scale, y_train)
y_pred = clf.predict(X_test)
recall = recall_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
print('Feature set:', feature_list)
print('Target label: six_mth_active')
print('Logistic Regression Classifier')
print('-accuracy:', clf.score(X_test_scale, y_test))
print('-recall:', recall)
print('-precision:', precision)
print('-F1:', 2 * precision * recall / (precision + recall))
fpr, tpr, thresh = roc_curve(y_test, y_pred)
plt.plot(fpr, tpr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC curve for logistic regression classifier')
plt.show()
```

```python
# clf = XGBClassifier(
#     learning_rate=0.1,
#     max_depth=6,
#     n_estimators=200,
#     cv=5,
#     n_jobs=-1
# )

clf = XGBClassifier()

clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
recall = recall_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
print('Feature set:', feature_list)
print('Target label: six_mth_active')
print('Gradient Boosted Classifier')
print('-accuracy:', clf.score(X_test, y_test))
print('-recall:', recall)
print('-precision:', precision)
print('-F1:', 2 * precision * recall / (precision + recall))
fpr, tpr, thresh = roc_curve(y_test, y_pred)
plt.plot(fpr, tpr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC curve for gradient boosted classifier')
plt.show()
```

<b>It seems the Gradient Boosted Classifier with default hyperparams gives the best results.  Playing around with the hyperparams yields very little improvement or even a decrease in scores.  While recall is lower than Logistic Regression, we precision is much higher.  There is also a good balance between precision and recall according to the F1 score, and the ROC curve shows a much better chance than 50/50 to predict the correct label.  This means less users are mistakenly classified as active/inactive in 6 months.  Resources can be put to more efficient use in keeping active customers or giving inactive customers incentive to become active.</b>
